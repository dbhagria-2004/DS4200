### Part 3 - Basic data analysis with Altair (40 points)

We are going to use the same Weather data from Homework 1 but use Altair to create the data visualziation. 

##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
weather['weather'].unique()
```

```python
alt.Chart(weather).mark_boxplot().encode(alt.Y("temp_max:Q", title="max_temp"), alt.X("weather:N", title= "weather_type")).properties(title= "Distribution of Max temp by weather type", width=600, height=400)
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
colors = {
    "drizzle": "blue",
    "rain": "green",
    "snow": "purple",
    "sun": "orange",
    "fog": "red"
}

# Scatter plot
alt.Chart(weather).mark_circle(size=60, opacity=0.7).encode(
    x=alt.X("temp_min:Q", title="Min Temperature (C)"),
    y=alt.Y("temp_max:Q", title="Max Temperature (C)"),
    color=alt.Color("weather:N",
                    scale=alt.Scale(domain=list(colors.keys()),
                                    range=list(colors.values())),
                    title="Weather Type"),
    tooltip=["date:T", "temp_min:Q", "temp_max:Q", "weather:N"]
).properties(
    title="Relationship between Max and Min Temperature (Colored by Weather)",
    width=500,
    height=400
)
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
alt.Chart(weather).mark_point().encode(
    alt.Y('temp_max:Q'),
      alt.X('temp_min:Q'), 
      alt.Color('wind:N', title='Wind speed'), alt.Tooltip(['temp_min', 'temp_max', 'wind'])).facet(
          facet='weather', title='Weather_type'
          , columns= 3).properties(title="Scatter Plot of Max vs Min Temperature by Weather").interactive()
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python
weather['average'] = (weather['temp_max'] + weather['temp_min'])/2

line = alt.Chart(weather).mark_line().encode(alt.X('date:T', title="Date"), alt.Y('average:Q', title="Average_temp"))

points = alt.Chart(weather).mark_point().encode(alt.X('date:T', title='Date'), alt.Y('average:Q', title=" Average_temp"), tooltip=['date', 'temp_min', 'temp_max', 'average'])

(line+points).properties(title= "Trend of Average Temperature Over 4 Years", width=800, height=200)
```
