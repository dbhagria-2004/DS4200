```python
#1 First Visualization using Altair, I arranged the survival count cateogorized by Sex

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
titanic_data = sns.load_dataset('titanic')


alt.Chart(titanic_data).mark_bar().encode(
    x=alt.X('sex:N', title="Sex"),
    y=alt.Y('count():Q', title="Passenger Count"),
    color=alt.Color('survived:N', title="Survived"),
    tooltip=['sex', 'survived', 'count()']
).properties(
    title="Survival by Sex", height=200, width=100
).interactive()
```

```python
#2 Second Visualization using Altair, I arranged the survival count cateogorized by Class, the passengers were travelling in.

import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
titanic_data = sns.load_dataset('titanic')

chart = alt.Chart(titanic_data).mark_bar().encode(
    x=alt.X('class:N', title="Passenger Class"),
    y=alt.Y('count():Q', title="Passenger Count"),
    color=alt.Color('survived:N',
                    title="Survived"),
    tooltip=['class', 'survived', 'count()']  # Tooltip for interaction
).properties(
    title="Survival by Class", height=200, width=100
).interactive()

chart
```

```python
#3 Third Visualization using Altair, I arranged the survival count cateogorized by Age Group, the passengers were of.
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
titanic_data = sns.load_dataset('titanic')

bins = [0, 12, 20, 40, 60, 80]
labels = ["Child", "Teen", "Adult", "Middle Age", "Senior"]

# Assign age groups
titanic_data['age_group'] = pd.cut(titanic_data['age'], bins=bins, labels=labels)


alt.Chart(titanic_data).mark_bar().encode(
    x=alt.X('age_group:N', title="Age Group", sort=labels),
    y=alt.Y('count():Q', title="Passenger Count"),
    color=alt.Color('survived:N',
                    title="Survived"),
    tooltip=['age_group', 'survived', 'count()'] 
).properties(
    title="Survival by Age Group"
)
```
